# Plot Taylor diagram:
for simus already done, comparison seasonnality

```python
import xarray as xr
import numpy as np
from taylor_diagram import ModTaylorDiagram
import matplotlib.pyplot as plt
```

```python
diagram = ModTaylorDiagram()

cafe={'norm_std':0.729724,'norm_bias':-0.244905,'correl_coef':0.956822,'norm_rmse':0.36887,'name':'CAFE'}
vgpm={'norm_std':0.487506,'norm_bias':-0.292965,'correl_coef':0.959701,'norm_rmse':0.549492,'name':'VGPM'}
cbpm_behr={'norm_std':0.985261,'norm_bias':-0.006676,'correl_coef':0.999957,'norm_rmse':0.017404,'name':'CbPM Behrenfeld'}
cbpm_west={'norm_std':0.916096,'norm_bias':-0.045437,'correl_coef':0.999622,'norm_rmse':0.087929,'name':'CbPM Westberry'}

all_npp = [cafe, vgpm, cbpm_behr, cbpm_west]

for npp in all_npp:
    diagram.add_prediction(npp['norm_std'], npp['norm_bias'], npp['correl_coef'], npp['norm_rmse'], npp['name'], r"")
diagram.plot()
plt.title(
    "Diagramme de Taylor - Différents produits de Production Primaire"
)

plt.show()
```
